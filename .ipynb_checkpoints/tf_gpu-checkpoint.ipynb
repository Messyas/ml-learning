```python
# Importando as bibliotecas
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np

# Verificando se a GPU está disponível
print("GPU disponível: ", "Sim" if tf.config.list_physical_devices('GPU') else "Não")

# Carregando o dataset MNIST
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Normalizando os valores de pixel (0-255 para 0-1)
train_images = train_images / 255.0
test_images = test_images / 255.0

# Construindo o modelo
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compilando o modelo
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Treinando o modelo
history = model.fit(train_images, train_labels, epochs=5, batch_size=64, validation_split=0.2)

# Avaliando o modelo
test_loss, test_acc = model.evaluate(test_images, test_labels)
print("\nAcurácia no teste: ", test_acc)

# Prevendo
predictions = model.predict(test_images[:5])

# Exibindo as primeiras previsões
print("\nPrimeiras previsões: ", np.argmax(predictions, axis=1))
print("Labels reais: ", test_labels[:5])
```

```python
# Importando as bibliotecas
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt

# Verificando se a GPU está disponível
print("GPU disponível: ", "Sim" if tf.config.list_physical_devices('GPU') else "Não")

# Carregando o dataset MNIST
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Normalizando os valores de pixel (0-255 para 0-1)
train_images = train_images / 255.0
test_images = test_images / 255.0

# Construindo o modelo
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compilando o modelo
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Treinando o modelo
history = model.fit(train_images, train_labels, epochs=5, batch_size=64, validation_split=0.2)

# Avaliando o modelo
test_loss, test_acc = model.evaluate(test_images, test_labels)
print("\nAcurácia no teste: ", test_acc)

# Prevendo
predictions = model.predict(test_images[:5])
print("\nPrimeiras previsões: ", np.argmax(predictions, axis=1))
print("Labels reais: ", test_labels[:5])

# Plotando os resultados (curva de perda e acurácia)
history_dict = history.history

# Plot da curva de perda
plt.figure()
plt.plot(history_dict['loss'], label='Perda de treinamento')
plt.plot(history_dict['val_loss'], label='Perda de validação')
plt.title('Curva de Perda')
plt.xlabel('Épocas')
plt.ylabel('Perda')
plt.legend()
plt.grid(True)
plt.show()

# Plot da curva de acurácia
plt.figure()
plt.plot(history_dict['accuracy'], label='Acurácia de treinamento')
plt.plot(history_dict['val_accuracy'], label='Acurácia de validação')
plt.title('Curva de Acurácia')
plt.xlabel('Épocas')
plt.ylabel('Acurácia')
plt.legend()
plt.grid(True)
plt.show()
```
